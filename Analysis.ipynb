# Loading Packages and Functions

```python
import pandas as pd
import numpy as num
import altair as alt
import vegafusion
import matplotlib
from sklearn.model_selection import train_test_split
```

# Reading Data

```python
# Reading the dataset
df = pd.read_csv("https://osf.io/download/g72pq/", sep=',', on_bad_lines='skip', low_memory=False)

# Creating a list of columns related to questions to drop
columns_to_drop = [col for col in df.columns if col.startswith('q')]

# Dropping the question columns from the DataFrame
df = df.drop(columns_to_drop, axis=1)

# Dropping the unrelavante columns from the DataFrame
df = df.drop("elogit", axis=1)
df = df.drop("tests", axis=1)
df = df.iloc[:,4:]

# Splitting the DataFrame
train_df, test_df = train_test_split(df, test_size=0.3, random_state=123)

df
```

## Distributions

```python
alt.data_transformers.enable('vegafusion')

numeric_cols = train_df.select_dtypes("number").columns.tolist()
numeric_cols_dist = alt.Chart(train_df).mark_bar().encode(
    alt.X(alt.repeat(), type = "quantitative", bin = alt.Bin(maxbins = 20)),
    y ="count()",
).properties(
        width = 250,
        height = 150
).repeat(
    numeric_cols,
    columns = 3
)

numeric_cols_dist
```

```python
splom = alt.Chart(train_df).mark_point(opacity = 0.4, size = 10).encode(
    alt.X(alt.repeat('row'), type = 'quantitative'),
    alt.Y(alt.repeat('column'), type = 'quantitative')
).properties(
    width = 200,
    height = 140
).repeat(
    column = numeric_cols,
    row = numeric_cols
)
splom
```
