# Reading and cutting down the dataset

## Downloading dataset

```python
import pandas as pd
dataset = pd.read_csv("https://osf.io/download/g72pq/", sep=',', on_bad_lines='skip', low_memory=False)
dataset.head()
```

## Randomly sampling 30% from the dataset to keep the file size below 100mb

```python
sampled_dataset = dataset.sample(frac=0.3,random_state=123)
sampled_dataset.info()
file_path = 'data/sampled_dataset.csv'
sampled_dataset.to_csv(file_path, index=False)
```
